```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

num_records = 100  # Number of packages to simulate
data = []

for _ in range(num_records):
    record = {
        "timestamp": datetime.now() - timedelta(minutes=np.random.randint(0, 1440)),  # Time of last sensor read
        "package_id": f"PKG{np.random.randint(1000, 9999)}",  # RFID-based unique package ID
        "product_id": f"PROD{np.random.randint(100, 999)}",  # Simulated product identity via RFID
        "location_lat": round(np.random.uniform(10.0, 14.0), 6),  # GPS Latitude
        "location_long": round(np.random.uniform(120.0, 125.0), 6),  # GPS Longitude
        "temperature_C": round(np.random.uniform(2.0, 8.0), 1),  # Temperature sensor for cold chain
        "status": np.random.choice(["In Transit", "Out for Delivery", "Delivered", "Delayed"]),
        "tamper_detected": np.random.choice([True, False], p=[0.05, 0.95])  # 5% chance of tampering
    }
    data.append(record)

# Create DataFrame
df = pd.DataFrame(data)

#  Save dataset to CSV
df.to_csv("iot_data.csv", index=False)

# Preview
df.head()
```
